## Multi-Model Anomaly Detection

To improve robustness, three anomaly detection models were used:

- Isolation Forest
- Local Outlier Factor (LOF)
- Robust Covariance (Elliptic Envelope)

An ensemble voting mechanism was implemented where a data point is labeled as anomalous only if at least 2 out of 3 models agree.

This reduces false positives and increases reliability of anomaly detection.

```python
from sklearn.ensemble import IsolationForest

iso_advanced_model = IsolationForest(
    contamination=0.02,
    random_state=42
)

iso_pred = iso_advanced_model.fit_predict(X_scaled)

df.loc[X.index, 'iso_advanced'] = iso_pred
df['iso_advanced'] = df['iso_advanced'].map({1:0, -1:1})
```

```python
from sklearn.ensemble import IsolationForest

iso = IsolationForest(
    contamination=0.01,   # 1% anomalies
    random_state=42
)

data['iso_label'] = iso.fit_predict(
    data[['electricity', 'rolling_mean_24', 'rolling_std_24', 'lag_1', 'lag_24']]
)

# Convert (-1 anomaly, 1 normal) → (1 anomaly, 0 normal)
data['iso_label'] = data['iso_label'].map({1:0, -1:1})

print(data['iso_label'].value_counts())
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(
    contamination=0.01
)

data['lof_label'] = lof.fit_predict(
    data[['electricity', 'rolling_mean_24', 'rolling_std_24', 'lag_1', 'lag_24']]
)

data['lof_label'] = data['lof_label'].map({1:0, -1:1})

print(data['lof_label'].value_counts())
```

```python
median = data['electricity'].median()
mad = (data['electricity'] - median).abs().median()

data['robust_z'] = 0.6745 * (data['electricity'] - median) / mad

data['robust_label'] = (data['robust_z'].abs() > 3).astype(int)

print(data['robust_label'].value_counts())
```

```python
data['ensemble'] = (
    data[['iso_label', 'lof_label', 'robust_label']]
    .sum(axis=1) >= 2
).astype(int)

print(data['ensemble'].value_counts())
```
